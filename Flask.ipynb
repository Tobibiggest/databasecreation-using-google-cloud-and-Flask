```python
from flask import Flask, render_template, request, redirect
from google.cloud import bigquery
```

```python
app = Flask(__name__)

# Initialize BigQuery client
client = bigquery.Client()

# Define the dataset and table names
dataset_id = 'ChurchNewcomersDB'
newcomers_table_id = 'Newcomers'
```

```python
# Define the route for the form
@app.route('/', methods=['GET', 'POST'])
def index():
    if request.method == 'POST':
        # Get form data
        firstname = request.form['firstname']
        lastname = request.form['lastname']
        gender = request.form['gender']
        age = int(request.form['age'])
        email = request.form['email']
        phone = request.form['phone']
        address = request.form['address']
        city = request.form['city']
        state = request.form['state']
        zipcode = request.form['zipcode']
        date_joined = request.form['date_joined']

        # Insert data into BigQuery
        insert_data(firstname, lastname, gender, age, email, phone, address, city, state, zipcode, date_joined)

        # Redirect to a thank you page
        return redirect('/thankyou')

    return render_template('index.html')
```

```python
# Define the route for the thank you page
@app.route('/thankyou')
def thankyou():
    return render_template('thankyou.html')

# Function to insert data into BigQuery
def insert_data(firstname, lastname, gender, age, email, phone, address, city, state, zipcode, date_joined):
    # Construct SQL query
    query = f"""
        INSERT INTO `{dataset_id}.{newcomers_table_id}` 
        (FirstName, LastName, Gender, Age, Email, Phone, Address, City, State, ZipCode, DateJoined) 
        VALUES 
        ('{firstname}', '{lastname}', '{gender}', {age}, '{email}', '{phone}', '{address}', '{city}', '{state}', '{zipcode}', '{date_joined}')
    """

    # Execute the query
    query_job = client.query(query)
    query_job.result()

if __name__ == '__main__':
    app.run(debug=True)
```
